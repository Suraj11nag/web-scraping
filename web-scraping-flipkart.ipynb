```python
links_list = []
```

```python
for link in links:
    links_list.append(link.get('href'))
```

```python
d = {"title": [], "price": [],  "average_rating": [], "rating": [], "review": []}
```

```python
# Function to extract Product Title
def get_title(soup):
    try:
        # Find the title element and extract its text
        title = soup.find("span", attrs={'class': 'VU-ZEz'}).text.strip()
    except AttributeError:
        title = ""
        print("Title not found")  # Optional: For debugging purposes
    return title

# Function to extract Product Price
def get_price(soup):

    try:
        price = soup.find("div", attrs={'class':'Nx9bqj CxhGGd'}).text.strip()

    except AttributeError:
            price = ""

    return price

# Function to extract Product Rating
def get_rating(soup):
    try:
        rating = soup.find("span", attrs={'class': 'Wphh3N'}).text.strip().split('&')[0].strip().split()[0]
    except AttributeError:
        rating = ""
        print("Rating not found")  # Debugging line
    return rating

# Function to extract Number of User Reviews
def get_review_count(soup):
    try:
        review = soup.find("span", attrs={'class':'Wphh3N'}).text.strip().split('&')[1].strip().split()[0]

    except AttributeError:
        review = ""	

    return review

# Function to extract average rating
def get_average_ratings(soup):
    try:
        average_rating = soup.find("div", attrs={'class': 'XQDdHH'}).text.strip()
    except AttributeError:
        average_rating = ""
        print("Average rating not found")  # Debugging line
    return average_rating
```

```python
for link in links_list:
        new_webpage = requests.get("https://www.flipkart.com"+link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['average_rating'].append(get_average_ratings(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['review'].append(get_review_count(new_soup))
```

```python
df = pd.DataFrame(d)
```
